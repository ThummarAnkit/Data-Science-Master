#### Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
plt.figure(figsize=(10,10))
sns.boxplot(x=df.Category,y=df.Rating,data=df,palette='Set2')
plt.title('App rating by category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```

As you can see business, communication, dating, finance, game, family, tools productivity has most outliers so we can say that these are the categories that affect the ratings.

#### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
category = pd.DataFrame(df['Category'].value_counts())
category.rename(columns={'Category':'Count'},inplace=True)
```

```python
category
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=category.index[:10],y='Count',data=category[:10],palette='hls')
plt.title('Top 10 app categories')
plt.xticks(rotation=90)
plt.show()
```
