# Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

__Ans. :__<br>

The __Probability Mass Function (PMF)__ and __Probability Density Function (PDF)__ are two important concepts in probability theory that are used to describe the probability distribution of a random variable.

The __Probability Mass Function (PMF)__ is used for discrete random variables, which take on a finite or countably infinite number of possible values. The PMF assigns a probability to each possible value of the random variable, such that the sum of all the probabilities is equal to 1. In other words, the PMF gives the probability that a discrete random variable takes a specific value.

For example, consider a fair six-sided die. The PMF of the die can be defined as:

P(X = 1) = 1/6<br>
P(X = 2) = 1/6<br>
P(X = 3) = 1/6<br>
P(X = 4) = 1/6<br>
P(X = 5) = 1/6<br>
P(X = 6) = 1/6<br>

where X is the random variable representing the outcome of a single roll of the die. Each value of the PMF corresponds to the probability of rolling a specific number on the die.

The __Probability Density Function (PDF)__, on the other hand, is used for continuous random variables, which take on a continuous range of possible values. The PDF assigns a probability density to each possible value of the random variable, such that the total area under the PDF curve is equal to 1. In other words, the PDF gives the probability that a continuous random variable falls within a certain range of values.

For example, consider the normal distribution with mean μ = 0 and standard deviation σ = 1. The PDF of this distribution is given by:

#### f(x) = (1/√(2πσ^2)) * exp(-(x-μ)^2/(2σ^2))

where x is a continuous random variable. The PDF assigns a probability density to each possible value of x, and the probability of x falling within a certain range is equal to the area under the PDF curve between those values. For example, the probability of x falling between -1 and 1 can be calculated by integrating the PDF over that range:

#### P(-1 ≤ X ≤ 1) = ∫(-1 to 1) f(x) dx

The result of this integration gives the probability that x falls within the range of -1 to 1.

#### F(x) = P(X ≤ x)

# Q7: Explain uniform Distribution with an example.

Uniform distribution is a continuous probability distribution that models events where all outcomes are equally likely over a specified interval. The probability density function (PDF) of a uniform distribution is a constant over the interval and zero outside the interval.

For example, suppose we roll a fair six-sided die. The outcomes of the roll are equally likely to be any of the numbers 1, 2, 3, 4, 5, or 6. We can model this as a uniform distribution over the interval [1, 6], where each outcome has an equal probability of 1/6.

The PDF of a uniform distribution with interval [a, b] is:

#### f(x) = 1 / (b - a) if a <= x <= b
0 otherwise

The mean and variance of a uniform distribution are:

#### mean = (a + b) / 2
#### variance = (b - a)^2 / 12

Uniform distributions are commonly used in simulations, games, and other situations where outcomes are equally likely over a range of values.

# Q8: What is the z score? State the importance of the z score.

__Ans. :__<br>

A z-score, also known as a standard score, is a measure of how many standard deviations an observation or data point is from the mean of a distribution. It is calculated as:

#### z = (x - mean) / standard deviation

where x is the observation, mean is the mean of the distribution, and standard deviation is the standard deviation of the distribution.

The z-score is important because it allows us to compare observations from different distributions that may have different units or scales. By standardizing the observations using the z-score, we can compare them on a common scale and make meaningful comparisons.

# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

__Ans. :__<br>

The __Central Limit Theorem (CLT)__ is a fundamental theorem in probability theory that states that, under certain conditions, the sum or mean of a large number of independent and identically distributed random variables will be approximately normally distributed, regardless of the underlying distribution of the individual random variables.

The conditions required for the Central Limit Theorem to hold are:

1. The sample size is large (usually at least 30).

2. The sample is drawn from a population with a finite mean and finite variance.

3. The samples are drawn independently from each other.

The __significance of the Central Limit Theorem__ is that it provides a way to estimate the distribution of a population parameter from a sample, even if we do not know the underlying distribution of the population. This is important in statistics, where we often want to estimate the mean or variance of a population based on a sample of data.

# Q10: State the assumptions of the Central Limit Theorem.

__Ans. :__<br>

The Central Limit Theorem (CLT) makes several assumptions in order to hold true:

__1. Independence:__ The random variables being sampled must be independent of each other.

__2. Sample Size:__ The sample size should be sufficiently large. Typically, this means at least 30 observations.

__3. Random Sampling:__ The observations must be sampled randomly from the population.

__4. Finite Population:__ The population being sampled from should have a finite mean and a finite variance.

__5. Identical Distribution:__ The random variables should be identically distributed.

__6. No Extreme Values:__ The population should not contain any extreme values, outliers, or skewness that may influence the results.
